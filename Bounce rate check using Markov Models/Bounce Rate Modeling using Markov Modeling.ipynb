```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('site_data.csv', header=None, names=['last_page_id', 'next_page_id'])
```

```python
# Data 
# last_page_id = current_page_id - 1
# ==> start pages have last_page_id = -1
# B - Bounce
# C - Close
```

```python
df.head()
```

```python
row_sums = dict(df['last_page_id'].value_counts())
df = df.groupby(['last_page_id', 'next_page_id']).size().reset_index(name='count')
df['page_val_count'] = df['last_page_id'].apply(lambda x: row_sums.get(x))
df['transition_probability'] = df['count']/df['page_val_count']
```

```python
df.head()
```

```python
# Initial state distribution
# that is last_page_id = -1
df[df['last_page_id'] == -1].sort_values('transition_probability', ascending=False)
```

```python
# Inference
# most likely users start with page 9
```

```python
# Initial state distribution
# that is last_page_id = -1
df[df['next_page_id'] == 'B'].sort_values('transition_probability', ascending=False)
```

```python
# Inference 
# Page 9 has a high bounce rate
```
